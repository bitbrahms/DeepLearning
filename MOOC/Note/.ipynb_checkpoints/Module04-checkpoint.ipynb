```python
def find_person(dict_users, strU):
     if strU in dict_users:
          return dict_users[strU]
     else:
          return 'Not Found'
if __name__ == "__main__":
     names = ['xiaoyun', 'xiaohong', 'xiaoteng', 'xiaoyi', 'xiaoyang']
     qq_num = [1,2,3,4,5]
     dict_users = dict(zip(names, qq_num))
     strU =  input()
     print(find_person(dict_users, strU))
```

```python
def countfeq(s):
    s_list = s.split('/')
    print(s_list)
    s_list = set(s_list)
    print(s_list)
    [s_list.remove(item) for item in s_list if item in '，。!！“”"']
    #print(s_list)
    s_list = set(s_list)
    data_dict={}
    for item in s_list:
        if item not in data_dict:
            data_dict[item] = 1
        else:
            data_dict[item] +=1
    return data_dict
if __name__ == "__main__":
    s = input()
    s_dict = countfeq(s)
    #print(s_dict)
    print(len(s_dict.keys()))
```
